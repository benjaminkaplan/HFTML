```python
from imports import *
from data_model2 import DataModel2
import pytz
from ml_model import MLModel
```

```python
d1 = DataModel2(pickle="pickles/AAPL_minute_long_120.pkl")
d1.xytrain(pct=1, shuffle=False)
```

```python
import datetime
boundary_dates = []
ny = pytz.timezone("America/New_York")
train_months = 3
test_months = 1
for i in range(12):
    boundary_dates.append(ny.localize(datetime.datetime(year=2016, month=i+1, day=1)))
    
min_date = boundary_dates[0]
```

```python
xtrains, ytrains = [], []
xtests, ytests = [], []
for i in range(train_months, 11):
    xtemp = d1.xtrain[(d1.xtrain.index.get_level_values('y_date') < boundary_dates[i])  & \
              (d1.xtrain.index.get_level_values('y_date') > max(min_date,boundary_dates[i-train_months]))]
    ytemp = d1.ytrain.loc[xtemp.index.get_level_values('y_date').unique()]
    xtrains.append(xtemp)
    ytrains.append(ytemp)
    
    xtemp = d1.xtrain[(d1.xtrain.index.get_level_values('y_date') < boundary_dates[i+1])  & \
              (d1.xtrain.index.get_level_values('y_date') > max(min_date,boundary_dates[i-test_months+1]))]
    ytemp = d1.ytrain.loc[xtemp.index.get_level_values('y_date').unique()]
    xtests.append(xtemp)
    ytests.append(ytemp)
    
```

```python
ml_models = []
data_models = []
i = 0
for i in range(7,len(xtrains)):
    idx = ytrains[i].index.get_level_values('y_date').unique().to_list()
    random.seed(1)
    random.shuffle(idx)
    x, y =  xtrains[i].loc[idx], ytrains[i].loc[idx]
    di = DataModel2(xtrain = x, ytrain = y, xtest = xtests[i], ytest = ytests[i], window=120)
    d1.save(path = "pickles/months/AAPL_minute_long_120_"+str(i)+".pkl")
    data_models.append(di)
    m = MLModel(data_model=di)
    m.create_ml_model_1()
    m.train_ml_model_1()
    ml_models.append(m)
    m.save(name = "months/AAPL_minute_long_120_"+str(i))
```
